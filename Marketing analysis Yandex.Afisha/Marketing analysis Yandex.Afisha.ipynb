## Продуктовые метрики

### Рассчет метрик DAU, WAU и MAU

Изучим данные пользовательской активности. Рассчитаем следующие метрики:

* DAU (от англ. daily active users) — количество уникальных пользователей в день;
* WAU (от англ. weekly active users) — количество уникальных пользователей в неделю;
* MAU (от англ. monthly active users) — количество уникальных пользователей в месяц.


Чтобы посчитать недельное и месячное использование, выделим во всех датасетах в отдельные столбцы  месяц и неделю, а также полную дату.

```python
# для датафрейма visits
visits['start_ts_month'] = visits['start_ts'].dt.month
visits['start_ts_week']  = visits['start_ts'].dt.week
visits['start_ts_date'] = visits['start_ts'].dt.date
```

Сгруппируем данные по уникальным пользователям и найдём среднее.

```python
dau_total = visits.groupby('start_ts_date').agg({'uid': 'nunique'}).mean()
wau_total = visits.groupby('start_ts_week').agg({'uid': 'nunique'}).mean()
mau_total = visits.groupby('start_ts_month').agg({'uid': 'nunique'}).mean()

print('DAU: ',int(dau_total))
print('WAU: ',int(wau_total))
print('MAU: ',int(mau_total))
```

Получается, что в среднем в день сайт Яндекс.Афиши посещает 907 уникальных пользователей, в неделю 5825, а в месяц 23228  уникальных пользователей.

Отобразим изменения метрик во времени на графиках.

```python
dau = visits.groupby('start_ts_date').agg({'uid': 'nunique'})
wau = visits.groupby('start_ts_week').agg({'uid': 'nunique'})
mau = visits.groupby('start_ts_month').agg({'uid': 'nunique'})

dau.plot(figsize=(15,4))
plt.title('DAU')
plt.ylabel('Количество уникальных посетителей')
plt.xlabel('День');
```

Количество уникальных пользователей в день с июня по ноябрь постепенно растет. При этом во второй половине ноября можно наблюдать стремительный рост пользователей. При среднем числе уникальных пользователей - 907, в один из дней ноября этот  показатель превысил 3000. Вероятно это связано с наступающими новогодними праздниками. Или на сайте в этот день была акция и скидки на билеты. Затем с января месяца число пользователей в день постепенно снижается и в конце марта число пользователей было близко к нулю. Возможно в тот день были технические работы и сайт был не доступен для пользователей.

```python
wau.plot(figsize=(15,4))
plt.title('WAU')
plt.ylabel('Количество уникальных посетителей')
plt.xlabel('Номер недели');
```

Количество уникальных пользователей в неделю имеет схожий общий тренд: с начала года идет постепенное снижение до конца лета и затем количество пользователей растет с пиком в конце ноября. При этом есть несколько недель в марте и мае, где число пользователей растёт. Вероятно это связано с праздниками и дополнительными выходными днями.

### Сколько раз за день пользователи заходят на сайт

Определим сколько раз за день пользователи в среднем заходят на сайт. Для этого  сгруппируем данные по дате и выведем общее количество посещений сайта и количество уникальных пользователей. Затем посчитаем нашу метрику и сохраним ее в столбце *sessions_per_user*

```python
sessions_by_date = visits.groupby('start_ts_date').agg({'uid': ('count', 'nunique')})
sessions_by_date.columns = ['n_sessions', 'n_users']
sessions_by_date['sessions_per_user'] = (sessions_by_date['n_sessions']/sessions_by_date['n_users'])
sessions_by_date['sessions_per_user'].mean()
```

Получается, что в среднем пользователи заходят на сайт 1.08 раз за день. Построим график, отражающий изменения метрики во времени.

```python
sessions_by_date.plot(y='sessions_per_user', figsize=(15,6));
plt.title('Количество посещений сайта одним пользователем в день');
plt.ylabel('Количество посещений');
plt.xlabel('День');
```

Чаще всего пользователи заходят несколько раз за день в ноябре-декабре, что вероятно также связано с поиском билетов на новогодние праздники.

Получается, что продолжительность типичной пользовательской сессии - 60 секунд. Вероятно большинство пользователей переходят на сайт по ссылке с внешнего  источника на конкретное мероприятие. 

### Расчёт Retention Rate

Рассчитаем показатель  Retention Rate, который показывает, сколько пользователей из когорты относительно их изначального числа вернулось в последующие периоды.

Возьмём дату, когда пользователь впервые зашел на сайт. Эти данные запишем в объект Series first_activity_date. Объединим его с исходным датафреймом visits.

```python
first_activity_date = visits.groupby(['uid'])['start_ts'].min()
first_activity_date.name = 'first_activity_date'
user_activity = visits.join(first_activity_date,on='uid')
```

Вызовем метод astype(): преобразуем значения столбцов к типу `datetime64[M]`.

```python
user_activity['first_visit_month'] = user_activity['first_activity_date'].astype('datetime64[M]')
user_activity['visit_month'] = user_activity['start_ts'].astype('datetime64[M]')
```

Для каждой строки датафрейма рассчитаем lifetime пользователя в рамках когорты. 

```python
user_activity['cohort_lifetime'] = user_activity['visit_month'] - user_activity['first_visit_month']
user_activity['cohort_lifetime'] = user_activity['cohort_lifetime'] / np.timedelta64(1,'M')
user_activity['cohort_lifetime'] = user_activity['cohort_lifetime'].round().astype('int')
```

Сгруппируем данные по когорте и lifetime. Посчитаем для каждой когорты количество активных пользователей на определённую «неделю жизни» и найдём исходное количество пользователей в когорте. Возьмём их число на нулевую неделю.

```python
cohorts = user_activity.groupby(['first_visit_month','cohort_lifetime']).agg({'uid':'nunique'}).reset_index()
initial_users_count = cohorts[cohorts['cohort_lifetime'] == 0][['first_visit_month','uid']]
initial_users_count = initial_users_count.rename(columns={'uid':'cohort_users'})
```

Объединим данные по когортам с исходным количеством пользователей в когорте.

```python
cohorts = cohorts.merge(initial_users_count,on='first_visit_month')
```

Рассчитаем Retention Rate. Разделим количество активных пользователей в каждую из недель на исходное число пользователей в когорте. Построим сводную таблицу и создадим тепловую карту.

```python
cohorts['retention'] = cohorts['uid']/cohorts['cohort_users']
retention_pivot = cohorts.query('cohort_lifetime > 0').pivot_table(index='first_visit_month',columns='cohort_lifetime',values='retention',aggfunc='sum')
retention_pivot
```

Retention Rate в разрезе когорот убывает с течением времени. При этом лучший показатель у июньской когорты, на 11-ом месяце коэффициент удержания составляет 4,5%. Причиной снижения пользовательской активности может быть масса факторов. ###  Изменение LTV на покупателя по когортам

Изучим как меняется LTV на покупателя по кагортам за 6 месяцев. Маржинальность сервиса - 100%.

Посмотрим на распределение дат заказов:

```python
orders['buy_ts'].describe() 
```

```python
costs['dt'].describe()
```

Диапазон дат расходов совпадает с датами заказов. Cократим период заказов на один день, чтобы данные совпадали.

```python
orders=orders.query('buy_ts<="2018-05-31 00:00:00"')
```

Выделим месяц из дат в таблице с  расходами и получим месяц первой покупки каждого покупателя.

```python
orders['order_month'] = orders['buy_ts'].astype('datetime64[M]')
costs['month'] = costs['dt'].astype('datetime64[M]') 

first_orders = orders.groupby('uid').agg({'order_month': 'min'}).reset_index()
first_orders.columns = ['uid', 'first_order_month']
first_orders.head()
```

Посчитаем количество новых покупателей (n_buyers) за каждый месяц и построим когорты.

```python
cohort_sizes = (
    first_orders.groupby('first_order_month')
    .agg({'uid': 'nunique'})
    .reset_index()
)
cohort_sizes.columns = ['first_order_month', 'n_buyers']

orders_new = pd.merge(orders,first_orders, on='uid')

cohorts = (
    orders_new.groupby(['first_order_month', 'order_month'])
    .agg({'revenue': 'sum'})
    .reset_index()
)

report = pd.merge(cohort_sizes, cohorts, on='first_order_month')

report.head()
```

Найдем валовую прибыль, умножив выручку на маржинальность и расчитаем возраст когорты и в результате посчитаем ltv.

```python
margin_rate = 1

report['gp'] = report['revenue'] * margin_rate
report['age'] = (
    report['order_month'] - report['first_order_month']
)/ np.timedelta64(1, 'M')
report['age'] = report['age'].round().astype('int')
```

```python
report['ltv'] = report['gp'] / report['n_buyers']

output = report.pivot_table(
    index='first_order_month', columns='age', values='ltv', aggfunc='mean'
)

output.fillna('') 
output = output.cumsum(axis=1)
output
```

Визуализируем данные при помощи тепловой карты.

```python
sns.set(style='white')
plt.figure(figsize=(12, 8))
plt.title('Накопительный LTV по когортам')
ax = sns.heatmap(output, annot=True, fmt='.2f', linewidths=1, linecolor='gray', vmin=0, vmax=14)
ax.set_yticklabels(output.index.strftime('%Y-%m-%d'))
plt.ylabel('Когорта')
plt.xlabel('Возраст когорты');
```

Самые стабильные показатели можно наблюдать у июньской когорты. Скорее всего это связано с тем, что у июньской когорты был и самый высокий коэффициент удержания пользователей. При этом у сентябрьской когорты на третий месяц резко растет LTV. Возможно в тот месяц была применена удачная рекламная стратегия к этой когорте и многие пользователи снова купили билеты на сайте.


Рассчитаем также средний LTV за 6 месяцев.

```python
sixmonth_mean = output[5].mean()
print('Средний LTV за 6 месяцев: {:.2f} у.е.'.format(sixmonth_mean))
```

### Вывод

Рассчитав метрики электронной коммерции мы определили, что пользователи в среднем проводят немного времени на сайте до совершения покупки - всего 60 секунд Это свидетельствует о том, что большинство пользователей приходят сайт, понимая билет на какое мероприятие они хотят купить. Также в среднем за полгода на одного клиента приходится 1,2 покупки. То есть часть клиентов возвращается и приобритает билеты через сайт повторно. Самый высокий средний чек бывает в марте и декабре. Вероятно высокий чек в эти месяцы связан с праздниками ( 8 марта и Новый год). Самый низкий чек наблюдается  в июне. Наиболее стабильные показатели  по LTV  у июньской когорты.
